```python
# sigmoid

def sigmoid(array):
    return 1/(1 + np.exp(-(array)))

a = np.arange(10)
sigmoid(a)
```

```python
# mean squared error

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
```

```python
actual
```

```python
predicted
```

```python
def mse(actual , predicted):
    return np.mean((actual - predicted) ** 2)

mse(actual,predicted)
```

Plotting graphs

```python
# plotting a 2d plot

# x = y
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
y = x
y

plt.plot(x,y)
```

```python
# y = x2

x = np.linspace(-10,10,100)
y = x ** 2

plt.plot(x,y)
```

```python
# sinx
x = np.linspace(-10,10,1000)
y = np.sin(x)

plt.plot(x,y)
```

```python
# y = xlogx

x = np.linspace(-10,10,100)
y = x * np.log(x)

plt.plot(x,y)
```

```python
# sigmoid graph

x = np.linspace(-10,10,100)
y = sigmoid(x)
# sigmoid function written above
plt.plot(x,y)
```
